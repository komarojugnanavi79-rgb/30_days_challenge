```python
#check whether the num is prime or not
def prime_num(n):
  if n <= 1:
    print("Not a prime ")
    return
  for i in range(2,n):
    if n % i==0:
      print("Not a prime")
      return
      break
  print("prime")
  return

n=int(input())
prime_num(n)
```

```python
#Print all prime numbers in a range.
n=int(input("Enter your range:"))
for i in range(2,n+1):
  for j in range(2,i):
    if i % j==0:
      break

  else:
      print(i)
```

```python
# gcd and hcf of 2 numbers
def gcd_hcf(a,b):
    hcf = 1

    for i in range(1, min(a,b)+1):
        if a % i == 0 and b % i == 0:
            hcf = i

    print(hcf)


a,b = list(map(int, input().split()))

gcd_hcf(a,b)
```
