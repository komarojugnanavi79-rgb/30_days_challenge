```python
#Take a 2D Array/List as input of size N × M and print all elements row-wise
print("Enter the number of rows and columns:")
n, m = map(int, input().split())

matrix = []

print(f"Enter {n} rows with {m} elements each:")

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print("\nMatrix elements are:")

for i in range(n):
    for j in range(m):
        print(matrix[i][j], end=" ")
    print()
```

```python
#Print the sum of each row
print("Enter the number of rows and columns:")
n, m = map(int, input().split())

matrix = []

print(f"Enter {n} rows with {m} elements each:")

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print("\nSum of each row:")

for i in range(n):
    total = 0
    for j in range(m):
        total += matrix[i][j]
    print(f"Row {i + 1} Sum = {total}")
```

```python
#Search for a key in the matrix
print("Enter the number of rows and columns:")
n, m = map(int, input().split())

matrix = []

print(f"Enter {n} rows with {m} elements each:")

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print("Enter the key to search:")
key = int(input())

found = False

for i in range(n):
    for j in range(m):
        if matrix[i][j] == key:
            print(f"\nKey found at Row {i}, Column {j}")
            found = True
            break
    if found:
        break

if not found:
    print("\nKey not found in the matrix.")
```

```python
#Leetcode -Ratate image(48)
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix)


        for i in range(n):
            for j in range(i+1, n):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]


        for row in matrix:
            row.reverse()
```

```python
#leetcode-spiral matrix(54)
class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        result = []
        if not matrix:
            return result

        top, bottom = 0, len(matrix) - 1
        left, right = 0, len(matrix[0]) - 1

        while top <= bottom and left <= right:
            # Traverse from Left → Right
            for j in range(left, right + 1):
                result.append(matrix[top][j])
            top += 1

            # Traverse from Top → Bottom
            for i in range(top, bottom + 1):
                result.append(matrix[i][right])
            right -= 1

            # Traverse from Right → Left (if still valid)
            if top <= bottom:
                for j in range(right, left - 1, -1):
                    result.append(matrix[bottom][j])
                bottom -= 1

            # Traverse from Bottom → Top (if still valid)
            if left <= right:
                for i in range(bottom, top - 1, -1):
                    result.append(matrix[i][left])
                left += 1

        return result
```
