```python
#selection sort
print("Enter the size of the array:")
n = int(input())

print(f"Enter {n} elements separated by spaces:")
arr = list(map(int, input().split()))

for i in range(n):
    min_index = i

    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    arr[i], arr[min_index] = arr[min_index], arr[i]

print("\nSorted Array:")
print(arr)
```

```python
#Bubble sort
print("Enter the size of the array:")
n = int(input())

print(f"Enter {n} elements separated by spaces:")
arr = list(map(int, input().split()))

for i in range(n):
    for j in range(0, n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print("\nSorted Array:")
print(arr)
```

```python
#Insertion Sort
print("Enter the size of the array:")
n = int(input())

print(f"Enter {n} elements separated by spaces:")
arr = list(map(int, input().split()))

for i in range(1, n):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key

print("\nSorted Array:")
print(arr)
```
