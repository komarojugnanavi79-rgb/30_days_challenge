```python
# Function to calculate the sum of first N natural numbers
def sum_of_n_numbers(n):
  sum=1
  for i in range(1,n+1):
    sum += i
  return sum
n= int(input("Enter a number:"))
print(sum_of_n_numbers(n))
```

```python
# Function to print the multiplication table of N
def multiplication_table(n):
  for i in range(1,11):
    print(f"{n} × {i} = {n*i}")

n=int(input("Enter a number:"))
multiplication_table(n)
```

```python
# Function to count total digits in an integer
def count_of_numbers(n):
  count=0
  while n > 0:
    n //= 10
    count += 1
  return count

n=int(input("Enter a number:"))
count_of_numbers(n)
```

```python
# Function to reverse the digits of an integer
def reverse_number(n):
  reverse=0
  while n > 0 :
    digit= n % 10
    reverse = reverse*10 + digit
    n //= 10
  return reverse
n=int(input("Enter a number:"))
print(reverse_number(n))
```

```python
def is_prime(n):
  if n <= 1:
    print("Not a prime number")
    return


  for i in range(2, n):
    if n % i == 0:
      print("Not a prime number")
      return
      break

  print("It is prime number")
  return

n=int(input("Enter a number:"))
is_prime(n)

```
